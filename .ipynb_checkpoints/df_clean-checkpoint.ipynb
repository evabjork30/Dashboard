```python
import pandas as pd
import numpy as np
from tabulate import tabulate
```

```python
df = pd.read_csv('Grades Dataset FINAL.csv', delimiter=';')
df
```

```python
df['Grade'].isna().sum()
```

```python
df['Grade'] = df['Grade'].astype(str).str.replace(',', '.').astype(float)
```

```python
df.nlargest(5, 'Grade')
```

```python
df.nsmallest(5, 'Grade')
```

```python
df.drop(df[df['Grade'] > 10].index, inplace=True)
df.nlargest(5, 'Grade')
```

```python
print(list(df.columns))
```

```python
df.nlargest(10, 'BirthYear')
```

```python
df.drop(df[df['BirthYear'] > 2010].index, inplace=True)
df.nlargest(5, 'BirthYear')
```

```python
df.nsmallest(5, 'BirthYear')
```

```python
count_under_1940 = (df['BirthYear'] < 1940).sum()
print(count_under_1940)
```

```python
df.nsmallest(5, 'Semester')
```

```python
count_under_2005 = (df['Semester'] < 20051).sum()
count_under_2005
```

```python
count_over_2005 = (df['Semester'] >= 20051).sum()
count_over_2005
```

```python
df.drop(df[df['Semester'] < 20051].index, inplace=True)
df.nsmallest(5, 'Semester')
```

```python
df.nsmallest(5, 'Semester')
```

```python
count_under_1940 = (df['BirthYear'] < 1930).sum()
print(count_under_1940)
```

```python
df.drop(df[df['BirthYear'] < 1930].index, inplace=True)
df.nsmallest(5, 'BirthYear')
```

```python
df['Credits'] = df['Credits'].astype(str).str.replace(',', '.').astype(float)
df.nlargest(10, 'Credits')
```
